# 列表推导器

```python
# 示例1：生成1到10的平方列表
squares = [i**2 for i in range(1, 10)]
squares
```

```python
# 示例2：过滤出1到20中的偶数
evens = [i for i in range(1,20) if i%2 ==0]
```

```python
# 示例3：将输入的字符串列表转换为整数列表
# 假设输入为"12345"
input_data = list(map(int, input().slipt()))
# 列表推导器实现
input_data = [int(i) for i in input().slipt()]
```

```python
# 示例4：生成一个nxn的零矩阵
n = 3
zero_matrix = [[n for n in range(n)] for n in range(n)]
```

```python
# 示例5：快速生成一个长度为n的斐波那契数列
# f(n) = f(n-1) + f(n-2)
n = 10
fib = [0, 1]
[fib.append(fib[-1] + fib[-2]) for _ in range(n-2)]
fib

# 2000000这个数直接等电脑寄吧
# fib = [0, 1]; [fib.append(fib[-1] + fib[-2]) for _ in range(2000000)]
```

```python
# 示例6：将二维列表展平为一维列表
# 定义一个3x3的二维矩阵（列表的列表）
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
# 使用列表推导式将二维矩阵展平为一维列表
# 遍历matrix中的每一行(row)，再遍历每一行中的每个元素(x)
# 将所有元素收集到一个新的列表flatten中
flatten = [x for row in matrix for x in row]
# a = []
# for row in matrix:
#     for i in row:
#         a.append(i)
flatten
```
