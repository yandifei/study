# 内置排序算法

```python
# 默认升序排序
nums = [5,3,8,6,7]
print(sorted(nums))
```

```python
# 降序排序
print(sorted(nums, reverse=True))
```

```python
word = ["apple", "banana", "kiwi", "cherry"]
sorted(word, key=len)
```

```python
# 列表中的sort()方法
# 原地修改列表，直接排序，不返回新的列表
num =  [5,3,8,6,7]
num.sort()
num
```
