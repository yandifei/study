# 数据转换
## 重名名轴索引

```python
df = pd.DataFrame({'A':['A0', 'A1', 'A2', 'A3'],
                 'B':['B0', 'B1', 'B2', 'B3'],
                 'C':['C0', 'C1', 'C2', 'C3']})
df
```

```python
# 使用rename方法对DataFrame的列名进行重命名
df.rename(columns={"A":"a","B":"b","C":"c"})
```

```python
# 使用rename方法将DataFrame的列名转换为小写
# str.lower是一个字符串方法，用于将字符串转换为小写
# axis="columns"指定了对列名进行操作，而不是对行索引
df.rename(str.lower, axis="columns")
```

```python
# 使用rename方法对DataFrame的索引进行重命名
# index参数用于指定要修改的索引映射关系
# {1:"a", 2:"b"}表示将索引值为1的改为"a"，索引值为2的改为"b"
df.rename(index={1:"a",2:"b"})
```

## 数据值替换

```python
df = pd.DataFrame({'姓名': ['李红','小明','马芳','国志'],
                   '性别': ['0','1','0','1'],
                   '籍贯': ['北京','甘肃','','上海']})
df
```

```python
df.replace("","广州", inplace=True)
df
```

```python
df.replace(["帝都","北京"],["魔都","上海"])
```

```python
df.replace({0:"女",1:"男"})
```
