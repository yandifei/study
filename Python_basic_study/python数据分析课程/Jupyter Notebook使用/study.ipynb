# 快捷键使用

基于原来的jupyter notebook，添加了一些常用的快捷键，方便使用，不一定和vscod的插件一致，但基本一致，如果有出入就会在下方做出标注
| 操作 | 快捷键 |
|------|--------|
| 运行当前单元 | `Ctrl+Enter` |
| 运行当前单元并向下插入新单元 | `Shift+Enter` |
| 运行当前单元并保持选中 | `Alt+Enter` |
| 运行所有单元 | `Ctrl+Shift+Enter` |
| 运行当前单元格及以上所有单元格 | `Ctrl+Shift+Alt+Enter` |
| 切换到代码模式 | `Y` |
| 切换到标记模式 | `M` |
| 切换到原始模式 | `R` |
| 在当前单元格上方插入单元格 | `A` |
| 在当前单元格下方插入单元格 | `B` |
| 删除当前单元格 | `DD`（连续按两次D） |
| 剪切单元格 | `X` |
| 复制单元格 | `C` |
| 粘贴单元格 | `V` |
| 撤销单元格操作 | `Z` |
| 重做单元格操作 | `Shift+Z` 或 `Ctrl+Y` |
| 上移当前单元格 | `Ctrl+Shift+↑` |
| 下移当前单元格 | `Ctrl+Shift+↓` |
| 合并所选单元格 | `Shift+M` |
| 拆分单元格 | `Ctrl+Shift+-` |
| 清除当前单元格输出 | `O` |
| 清除所有单元格输出 | `Ctrl+Shift+O` |
| 保存笔记本 | `Ctrl+S` |
| 打开命令面板 | `Ctrl+Shift+P` 或 `F1` |
| 切换行号 | `L` |
| 显示键盘快捷键参考 | `Ctrl+K Ctrl+R` 或 `Ctrl+Shift+R` |
| 进入编辑模式 | `Enter` |
| 退出编辑模式（命令模式） | `Esc` |
| 选择上方单元格 | `↑`（命令模式下） |
| 选择下方单元格 | `↓`（命令模式下） |
| 扩展选择上方单元格 | `Shift+↑` |
| 扩展选择下方单元格 | `Shift+↓` |
| 将光标移到下一个编辑区域 | `Tab`（编辑模式下） |
| 将光标移到上一个编辑区域 | `Shift+Tab`（编辑模式下） |
| 代码自动补全 | `Ctrl+Space` |
| 显示文档或提示 | `Ctrl+I` |
| 中断内核 | `Ctrl+Shift+Space` 或 `I I`（连续两次I） |
| 重启内核 | `Ctrl+Shift+Alt+Space` 或 `0 0`（连续两次0） |
| 重启内核并运行所有单元格 | `Ctrl+Shift+F5` |

# vscode使用jupyter
1. 在插件中搜索python下载安装
2. 在插件中搜索jupyter下载安装
3. 创建ipynb文件后就会自动激活对应的插件配置
4. 选择内核(其实就是选择python解释器)
5. 写一条代码(print("hello world"))并运行，如果运行成功就代表环境配置好了

# 内核问题
如果遇到了第一次运行代码发现一直不出结果(可执行代码块左边一直转圈圈)或右下角有vscode的通知提示重启内容中且一直没有反应则有可能是环境中库的问题：
此时直接进入对应的python虚拟环境然后安装下面的库就行了（安装后重启vscode）。
pip install jupyter-core -U
pip install ipykernel -U
pip install traitlets -U
pip install jupyter-core -U
pip install jupyter -U
