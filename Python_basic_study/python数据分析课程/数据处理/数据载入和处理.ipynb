## 空值和缺失值的处理

```python
from numpy import nan
ser_obj = pd.Series([1, None, nan])
ser_obj
```

```python
pd.insnull(ser_obj)
```

```python
pd.notnull(ser_obj)
```

```python
import numpy as np
df_obj = pd.DataFrame({"类别":['小说', '散文随笔', '青春文学', '传记'],
                       "书名":[np.nan, '《皮囊》', '《旅程结束时》', '《老舍自传》'],
                       "作者":["老舍", None, "张其鑫", "老舍"]})
df_obj
```

```python
df_obj.dropna()
```

```python
# 使用 fillna 方法填充 DataFrame 中的缺失值
# fillna() 是 pandas 中用于处理缺失值的方法
# 参数 "66" 表示将所有的缺失值（NaN）替换为字符串 "66"
df_obj.fillna("66")
```

```python
df_obj = pd.DataFrame({'A': [1, 2, 3, nan],
                       'B': [nan, 4, nan, 6],
                       'C': ['a', 7, 8, 9],
                       'D':[nan, 2, 3, nan]})
df_obj
```

```python
df_obj.fillna({"a":4, "B":5, "D":6})
```

```python
df_obj.fillna(method="ffill")
```

```python
df_obj.fillna(method="bfill")
```
