对“某超市公司五店销售数据.xlsx”文件进行以下操作，将源代码与运行结果以截图形式上传

一、数据分组
1. 读取文件，并显示头部的数据，默认为5，也可以自由设置参数
2. 按“门店名称”列进行分组，统计总销售额
3. 按“产品类别”列进行分组，统计总利润额

二、数据聚合运算
4. 按“门店名称”列进行分组，分别统计利润额的总和和中位数，以及销售额的最大值和最小值
5. 按“产品类别”列进行分组，分别统计销售额的最大值、最小值、平均值、中位数

三、透视表和交叉表
6. 利用pivot_table创建透视表，分组后行索引为“产品类别”，列标签为“门店名称”，计算列为销售额
7. 利用cross_tab创建交叉表，分组后行索引为“产品类别”，列标签为“门店名称”，计算销售额的均值

```python
import pandas as pd
import numpy as np
```

```python
# 一、数据分组
# 1. 读取文件，并显示头部的数据，默认为5，也可以自由设置参数

# 读取文件
sales_data = pd.read_excel('某超市公司五店销售数据.xlsx')
# 显示头部的数据
sales_data.head()
```

```python

# 2. 按“门店名称”列进行分组，统计总销售额
sum_ales = sales_data.groupby('门店名称')['销售额'].sum()
sum_ales
```

```python
# 3. 按“产品类别”列进行分组，统计总利润额
sum_revenue = sales_data.groupby('产品类别')['利润额'].sum()
sum_revenue
```

```python
# 二、数据聚合运算
# 4. 按“门店名称”列进行分组，分别统计利润额的总和和中位数，以及销售额的最大值和最小值
grouped_data = sales_data.groupby('门店名称')
aggregated_data = grouped_data.agg({'利润额': ['sum', 'median'], '销售额': ['max', 'min']})
aggregated_data
```

```python
# 5. 按“产品类别”列进行分组，分别统计销售额的最大值、最小值、平均值、中位数
grouped_data = sales_data.groupby('产品类别')
aggregated_data = grouped_data.agg({'销售额': ['max', 'min', 'mean', 'median']})
aggregated_data
```

```python
# 三、透视表和交叉表
# 6. 利用pivot_table创建透视表，分组后行索引为“产品类别”，列标签为“门店名称”，计算列为销售额
pivot_table = sales_data.pivot_table(index='产品类别', columns='门店名称', values='销售额')
pivot_table
```

```python
# 7. 利用cross_tab创建交叉表，分组后行索引为“产品类别”，列标签为“门店名称”，计算销售额的均值
# 这里我得要注意crosstab是pd才有的，目前的sales_data数据格式DataFarme其实是没有crosstab方法的，如果硬来需要整体进行数据转换
cross_table = pd.crosstab(
    index=sales_data['产品类别'],
    columns=sales_data['门店名称'],
    values=sales_data['销售额'],
    aggfunc=np.mean
)
cross_table
```
