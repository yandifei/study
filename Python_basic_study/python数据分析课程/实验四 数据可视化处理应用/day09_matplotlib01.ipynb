### 6.2.2 通过subplot()函数创建单个子图

```python
nums = np.arange(0, 101)    # 生成0~100的数组
# 分成2*2的矩阵区域，占用编号为1的区域，即第1行第1列的子图
plt.subplot(221)
# 在选中的子图上作图
plt.plot(nums, nums)
# 分成2*2的矩阵区域，占用编号为2的区域，即第1行第2列的子图
plt.subplot(222)
# 在选中的子图上作图
plt.plot(nums, -nums)
# 分成2*1的矩阵区域，占用编号为2的区域，即第2行的子图
plt.subplot(212)
# 在选中的子图上作图
plt.plot(nums, nums**2)
# 在本机上显示图形
plt.show()
```

### 6.2.3 通过subplots()函数创建多个子图

```python
# 生成包含1～100之间所有整数的数组
nums = np.arange(1, 101)
# 分成2*2的矩阵区域，返回子图数组axes
fig, axes = plt.subplots(2, 2)
ax1 = axes[0, 0]  # 根据索引［0，0］从Axes对象数组中获取第1个子图
ax2 = axes[0, 1]   # 根据索引［0，1］从Axes对象数组中获取第2个子图
ax3 = axes[1, 0]   # 根据索引［1，0］从Axes对象数组中获取第3个子图
ax4 = axes[1, 1]   # 根据索引［1，1］从Axes对象数组中获取第4个子图
# 在选中的子图上作图
ax1.plot(nums, nums)
ax2.plot(nums, -nums)
ax3.plot(nums, nums**2)
ax4.plot(nums, np.log(nums))
plt.show()
```

## 案例
### 统计某班级40名学生的期末考试成绩（满分100分），成绩分布如下（分组统计）：60分以下5人、60-70分8人、70-80分12人、80-90分10人、90-100分5人。
### 核心数据：成绩等级（不及格、及格、良好、优秀、卓越）、对应人数、对应占比。
### 分析目标
### 1、创建原始数据（成绩等级、对应人数）
### 2、计算对应占比
### 3、用柱状图直观展示各成绩等级的学生人数，对比不同等级的人数差异；
### 4、用饼图展示各成绩等级的人数占比，清晰呈现班级成绩分布结构；
### 5、将上述两张图绘制到同一张画布上并保存到本地
### 6、输出核心数据
### 7、通过图表解读，分析班级成绩整体情况（如是否偏科、优秀率/不及格率是否合理）。

```python
# 1. 导入所需库
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 2. 设置中文字体（避免图表中文乱码）
plt.rcParams['font.sans-serif'] = ['SimHei']  # 适配中文
plt.rcParams['axes.unicode_minus'] = False    # 适配负号

# 3. 整理数据（模拟班级成绩统计数据）
data = {
    '成绩等级': ['不及格(60以下)', '及格(60-70)', '良好(70-80)', '优秀(80-90)', '卓越(90-100)'],
    '人数': [5, 8, 12, 10, 5]
}
df = pd.DataFrame(data)

# 4. 计算各等级人数占比（用于饼图）
df['占比(%)'] = (df['人数'] / df['人数'].sum() * 100).round(2)

# 5. 绘制柱状图（展示各等级人数）
plt.figure(figsize=(12, 6))  # 设置图表大小

# 子图1：柱状图
plt.subplot(1, 2, 1)
bars = plt.bar(df['成绩等级'], df['人数'], color=['red', 'orange', 'yellow', 'green', 'blue'])
plt.title('各成绩等级学生人数柱状图', fontsize=14, fontweight='bold')
plt.xlabel('成绩等级', fontsize=12)
plt.ylabel('学生人数', fontsize=12)
plt.xticks(rotation=45)  # 旋转x轴标签，避免重叠

# 子图2：饼图（展示各等级人数占比）
plt.subplot(1, 2, 2)
# 饼图参数：数据、标签、占比显示、颜色、分离效果（突出优秀等级）
explode = (0, 0, 0, 0.1, 0)  # 优秀等级轻微分离
plt.pie(df['人数'], labels=df['成绩等级'], 
        colors=['red', 'orange', 'yellow', 'green', 'blue'],
        explode=explode, shadow=True, startangle=90)
plt.title('各成绩等级学生人数占比饼图')

# 保存图表（可直接保存到本地）
plt.savefig('学生成绩分析图表.png', dpi=300, bbox_inches='tight')
plt.show()

# 6. 输出数据汇总表
print(df)
```

## 6.5	案例—画图分析某年旅游景点数据

### 6.5.3	功能实现

```python
import pandas as pd
import numpy as np
# 使用read_csv()方法进行读取
scenery_file_path = open(r'C:/Users/admin/Desktop/风景名胜区.csv')
scenery_data = pd.read_csv(scenery_file_path)
scenery_data
```

```python
# 计算‘总面积(平方公里)’的平均数，并保留一位小数
area = float("{:.1f}".format(
scenery_data['总面积(平方公里)'].mean()))
# 计算‘游客量(万人次)’平均数，并保留一位小数
tourist = float("{:.1}".format(
scenery_data['游客量(万人次)'].mean()))
# 将上述计算的平均值，使用fillna()函数，字典映射的形式进行填充
values = {"总面积(平方公里)":area,"游客量(万人次)":tourist}
scenery_data = scenery_data.fillna(value=values)
scenery_data.head
```

```python
 # 通过groupby()函数按“省份”一列拆分scenery_data
data = scenery_data.groupby("省份")
# 显示“河北”分组的数据
hebei_scenery = dict([x for x in data])['河北']
hebei_scenery
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif']=['SimHei']  # 正常显示中文标签
plt.rcParams['axes.unicode_minus']=False    # 正常显示负号
area = hebei_scenery['总面积(平方公里)'].values
tourist = hebei_scenery['游客量(万人次)'].values
# 设置尺寸
plt.figure(figsize=(12, 6))
x_num = range(0, len(area))
x_dis = [i + 0.3 for i in x_num]
plt.bar(x_num, area, color='g', width=.3, label='总面积')
plt.bar(x_dis, tourist, color='r', width=.3, label='游客量')
plt.ylabel('单位：平方公里/万人次')
plt.title('河北景点面积及游客数量')
# 设置图例
plt.legend(loc='upper right')
plt.xticks(range(0, 10),['苍岩山', '嶂石岩', '西柏坡-天桂山', 
'秦皇岛北戴河','响堂山','娲皇宫','太行大峡谷',
'崆山白云洞','野三坡','承德避暑山庄外八庙'])
plt.show()
```

```python
import matplotlib.pyplot as plt
every_scenery = hebei_scenery['游客量(万人次)'].values
all_scenery = hebei_scenery['游客量(万人次)'].sum()
# 计算每个景点游客所占百分比  保留两位小数
percentage = (every_scenery/all_scenery)*100
np.set_printoptions(precision=2)
labels  = ['苍岩山', '嶂石岩', '西柏坡-天桂山', 
            '秦皇岛北戴河','响堂山','娲皇宫','太行大峡谷',
            '崆山白云洞','野三坡','承德避暑山庄外八庙']
plt.axes(aspect=1)
plt.pie(x= percentage, labels=labels, autopct='%3.2f %%',shadow=True, labeldistance=1.2, startangle = 90,pctdistance = 0.7)
plt.legend(loc='upper left')
plt.show()
```
