使用 Matplotlib 绘制柱状图的 Python 代码，展示 4 种水果的供应量并为每种水果指定不同颜色。

```python
# 导入matplotlib库中的pyplot模块，并将其简称为plt，用于绘图
import matplotlib.pyplot as plt
# 在Jupyter Notebook或类似环境中使用此命令，可以直接在notebook中显示图形，而不是弹出新窗口
%matplotlib inline
# 设置中文字体
# 用来正常显示中文标签
plt.rcParams['font.sans-serif'] = ['SimHei']
# 用来正常显示负号
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 水果名称和对应的供应量
fruits = ['苹果', '香蕉', '橙子', '葡萄']
# 单位：吨
supply = [150, 200, 120, 180]

# 指定不同的颜色
colors = ['red', 'yellow', 'orange', 'purple']

# 创建柱状图
plt.bar(fruits, supply, color=colors, width=0.6)

# 添加标题和坐标轴标签
plt.title('4种水果供应量统计')
plt.xlabel('水果种类')
plt.ylabel('供应量 (吨)')

# 在柱子上方显示数值
for i, v in enumerate(supply):
    plt.text(i, v + 5, str(v), ha='center', va='bottom')

# 显示图形
plt.show()
```
