```python
arr2 = np.arange(20).reshape(4,5)
arr2
```

```python
# 将数组扁平化，即将原本可能是二维、三维甚至更高维度的数组，拉伸成一条线，变成一个一维数组
arr.ravel()
```

```python
arr3 = arr2*2
arr3
```

```python
# 水平堆叠 = 左右拼接。它的作用是将数组在水平方向（按列）顺序拼接。
np.hstack((arr2,arr3))
```

```python
# 垂直堆叠 = 上下拼接。它的作用是将数组在垂直方向（按行）顺序拼接。
np.vstack((arr2,arr3))
```

```python
# 水平分割 = 左右切（按列切）。它的作用是将数组在水平方向切分成若干份。
np.hsplit(arr3,5)
```

```python
# 垂直分割 = 上下切（按行切）。它的作用是将数组在垂直方向切分成若干份。
np.vsplit(arr3,4)
```

```python
# 数组的转置。简单来说，就是把数组的行和列互换，就像矩阵转置一样
arr3.transpose()
```

```python
arr4 = np.arange(12).reshape(3,4)
arr4
```

```python
# 交换数组的两个指定轴（维度）。代表第 0 个轴（通常指行）。1：代表第 1 个轴（通常指列）。
# 就是把第 0 轴和第 1 轴的位置互换。
arr5 = arr4.swapaxes(0,1)
arr5
```
