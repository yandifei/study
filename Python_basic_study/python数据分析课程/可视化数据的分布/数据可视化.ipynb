```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    '地区': [1, 2, 3, 4, 5, 6, 7],
    '销售额': [138,186, 226, 237, 278,356, 385]
}

df = pd.DataFrame(data)

p1 = df["地区"]
p2 = df["销售额"]

plt.plot(p1,p2,color='y',marker='d')
sns.despine(left=True,bottom=True)
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))

sns.set(style='darkgrid',font_scale=1.5)
ax1 = fig.add_subplot(2,2,1)
plt.plot(p1,p2,color='g',marker='o')

sns.set(style='whitegrid')
ax2 = fig.add_subplot(2,2,2)
plt.plot(p1,p2,color='m',marker='*')

sns.set(style='white')
ax3 = fig.add_subplot(2,2,3)
plt.plot(p1,p2,color='r',marker='D')

sns.set(style='dark')
ax4 = fig.add_subplot(2,2,4)
plt.plot(p1,p2,color='y',marker='d')

plt.show()
```

```python
fig,ax = plt.subplots(2,2,figsize=(10,7))

sns.histplot(tips['total_bill'],ax=ax[0,0],kde=True)
sns.histplot(tips['total_bill'],ax=ax[0,1],bins=20)
sns.kdeplot(tips['total_bill'],ax=ax[1,0],fill=True,color='r')
sns.rugplot(tips['total_bill'],ax=ax[1,1],color='g')
plt.show()
```

```python
# 关键步骤：加载数据集
tips = sns.load_dataset('tips')

# 散点图
fig = plt.figure(figsize=(12,8))

ax1 = fig.add_subplot(2,2,1)
plt.scatter(x=tips['day'], y=tips['tip'], c='r')

ax2 = fig.add_subplot(2,2,2)
sns.stripplot(x='day', y='tip', data=tips, color='m')

ax3 = fig.add_subplot(2,2,3)
sns.stripplot(x='day', y='tip', data=tips, color='c', jitter=True)

ax4 = fig.add_subplot(2,2,4)
sns.swarmplot(x='day', y='tip', data=tips, hue='sex', size=4)

plt.show()
```

```python
sns.boxplot(x='day', y='tip', data=tips, color='m')
plt.show()
```

```python
sns.set(style='whitegrid',palette='Accent')
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(2,2,1)
sns.violinplot(x='day',y='total_bill',hue='time',data=tips)

ax2 = fig.add_subplot(2,2,2)
sns.violinplot(x='day',y='total_bill',hue='sex',data=tips,split=True,inner='quartiles')
plt.show()
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')

fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(2,2,1)
sns.barplot(x='day',y='total_bill',hue='sex',data=tips,palette='Accent')

ax2 = fig.add_subplot(2,2,2)
sns.countplot(x='petal_width',data=iris,palette='husl')
plt.show()

iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.regplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker',data=tips,ci=None)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker',col='time',row='sex',data=tips)
plt.show()
```
