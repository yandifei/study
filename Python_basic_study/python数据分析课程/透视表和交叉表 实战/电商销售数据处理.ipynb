```python
import pandas as pd
import numpy as np
```

```python
sales = pd.read_csv('sales.csv', encoding='gbk')
sales
```

```python
sales.info()          # 查看字段类型及缺失情况
sales.head()          # 查看前5行
sales.describe()      # 数值列统计描述
```

```python
# 拿到订单金额、购买数量的平均值
# order_amount_mean = sales['订单金额'].mean()
# order_quantity_mean = sales['购买数量'].mean()
# # 订单金额、购买数量字段进行填充
# sales.fillna(value={'订单金额': order_amount_mean, '购买数量': order_quantity_mean}, inplace=True)
# sales

# 删除所有含有缺失值的行
sales = sales.dropna()
sales
```

```python
# 检查是否有重复值
sales.duplicated()
```

```python
# 删除重复值（订单编号 才是唯一的键值）
sales = sales.drop_duplicates(subset=['订单编号'])
sales
```

```python
# 使用pandas库的DataFrame和matplotlib库进行数据可视化
sales.boxplot()  # 绘制箱线图
plt.show()  # 显示图形
```

```python
# # 检测订单金额异常值
# Q1 = sales['订单金额'].quantile(0.25)
# Q3 = sales['订单金额'].quantile(0.75)
# IQR = Q3 - Q1
# lower = Q1 - 1.5 * IQR
# upper = Q3 + 1.5 * IQR
# outlier_mask = (sales['订单金额'] < lower) | (sales['订单金额'] > upper)
# sales.loc[outlier_mask, '订单金额'] = sales['订单金额'].mean()

# # 同理处理购买数量
# Q1 = sales['购买数量'].quantile(0.25)
# Q3 = sales['购买数量'].quantile(0.75)
# IQR = Q3 - Q1
# lower = Q1 - 1.5 * IQR
# upper = Q3 + 1.5 * IQR
# outlier_mask = (sales['购买数量'] < lower) | (sales['购买数量'] > upper)
# sales.loc[outlier_mask, '购买数量'] = sales['购买数量'].mean().round()

# 这里是删除异常值而不是填充（金额不能造假）

# 删除订单金额异常值
Q1 = sales['订单金额'].quantile(0.25)
Q3 = sales['订单金额'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outlier_mask = (sales['订单金额'] < lower) | (sales['订单金额'] > upper)

# 先看一下会删掉多少
print(f"金额异常值数量：{outlier_mask.sum()}")

# 删除这些行
sales = sales[~outlier_mask]

# 购买数量同理，可以直接删除
Q1_qty = sales['购买数量'].quantile(0.25)
Q3_qty = sales['购买数量'].quantile(0.75)
IQR_qty = Q3_qty - Q1_qty
lower_qty = Q1_qty - 1.5 * IQR_qty
upper_qty = Q3_qty + 1.5 * IQR_qty
qty_outlier_mask = (sales['购买数量'] < lower_qty) | (sales['购买数量'] > upper_qty)

print(f"数量异常值数量：{qty_outlier_mask.sum()}")
sales = sales[~qty_outlier_mask]

sales
```

```python
# 将订单时间字段转换为日期时间格式
sales['订单时间'] = pd.to_datetime(sales['订单时间'], format="%H:%M:%S")
sales
```

```python
# 新增月份列
sales['月份'] = sales['订单时间'].dt.month
sales
```

```python
# 总销售额、总订单数、总销量
total_sales = sales['订单金额'].sum()
# 去重统计订单数
total_orders = sales['订单编号'].nunique()
total_quantity = sales['购买数量'].sum()
print(f"总销售额：{total_sales:,.2f} 元")
print(f"总订单数：{total_orders} 单")
print(f"总销量：{total_quantity} 件")
```

```python
# 客单价
avg_order_value = total_sales / total_orders
print(f"客单价：{avg_order_value:.2f} 元/单")

# 各商品类别销售额与销量
category_stats = sales.groupby('商品类别').agg(
    销售额=('订单金额', 'sum'),
    销量=('购买数量', 'sum')
).sort_values('销售额', ascending=False)

print("各商品类别销售情况")
print(category_stats)
```

```python
# 各地区销售额并排序
region_sales = sales.groupby('销售区域')['订单金额'].sum().sort_values(ascending=False)
print("各地区销售额排名")
print(region_sales)
```

```python
# 月度销售趋势（订单总销售额）
monthly_trend = sales.groupby('月份')['订单金额'].sum().sort_index()
print("月度销售额趋势")
print(monthly_trend)
```

```python
# 用户复购率（购买≥2次的用户）
# 每个用户的订单数（去重订单编号）
user_order_count = sales.groupby('用户ID')['订单编号'].nunique()
print(user_order_count)
repurchase_users = (user_order_count >= 2).sum()      # 购买≥2次的用户数
total_users = user_order_count.count()                # 总用户数
repurchase_rate = repurchase_users / total_users

print("用户复购分析")
print(f"总用户数：{total_users} 人")
print(f"复购用户数（购买≥2次）：{repurchase_users} 人")
print(f"复购率：{repurchase_rate:.2%}")
```
