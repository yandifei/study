```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'k1':['a','b','a','a','c','c','b','a','c','a','b','c'],
    'k2':['one','two','three','two','one','one','three','one','two','three','one','two'],
    'w':np.random.rand(12),
    'y':np.random.randn(12)    
})
df
```

```python
#  使用pivot_table方法创建数据透视表 index参数指定行索引
# 这里使用'k1'列作为行索引 columns参数指定列索引
# 这里使用'k2'列作为列索引 aggfunc参数指定聚合函数
# 这里使用求和函数"sum"对数据进行汇总
df.pivot_table(index='k1',columns='k2', aggfunc="sum")
```

```python
df.pivot_table(index='k1',columns='k2', value="w")
```

```python
# 使用pandas库中的crosstab函数创建交叉表
# 交叉表是一种用于分析分类变量之间关系的表格
# 参数df.k1和df.k2分别表示DataFrame中的两列分类数据
pd.crosstab(df.k1,df.k2)
```

```python
pd.crosstab(df.k1,df.k2, margins=True)
```
