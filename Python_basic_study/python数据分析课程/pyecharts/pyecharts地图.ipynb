```python
# 从 pyecharts 的图表模块中导入地图组件 Map
from pyecharts.charts import Map

# 准备地图所需的数据，格式为包含 (省份名称, 数值) 元组的列表
data = [('江苏省', 33100), ('广东省', 45600), ('浙江省', 22350), ('四川省', 15800), ('山东省', 23500)]

# 实例化一个 Map 地图对象
map = Map()

# 向地图对象中添加数据：指定系列名称为 'gdp(单位: 亿)'，传入数据列表 data，并指定地图类型为中国地图 'china'
map.add('gdp(单位: 亿)', data, 'china')

# 设置地图的标题
map.set_global_opts(
    title_opts=opts.TitleOpts(title="2026Q1季度我国各省GDP排名前五"),
    visualmap_opts=opts.VisualMapOpts(max_=50000)
)

# 渲染地图并保存为 HTML 文件
map.render('map.html')
map.render_notebook()
```
