```python
# 1. 导入正确的二维散点图组件和配置项
from pyecharts.charts import Scatter
from pyecharts import options as opts

# 2. 准备数据
week = ['周一', '周二', '周三', '周四', '周五', '周六', '周日']
data = [80, 15, 150, 49, 59, 97, 15]

# 3. 实例化散点图对象
s = Scatter()

# 4. 添加 X 轴和 Y 轴数据
s.add_xaxis(week)
s.add_yaxis('商家', data)

# 5. 设置全局配置项（注意：之前缺少的 opts. 已经补上）
s.set_global_opts(title_opts=opts.TitleOpts(title='一周营业额（万元）'))

# 6. 渲染并保存图表
s.render('scatter.html')
s.render_notebook()
```
