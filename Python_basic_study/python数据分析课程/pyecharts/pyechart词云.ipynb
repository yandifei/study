```python
# 从 pyecharts 的图表模块中导入词云图组件 WordCloud
from pyecharts.charts import WordCloud
from pyecharts import options as opts

# 准备词云图所需的数据，格式为一个包含多个 (词语, 词频/权重) 元组的列表
words = [
    ("竹升面", 5800), ("珠江", 6181),
    ("《广州日报》", 4386), ("猴赛雷", 3055),
    ("广东省博物馆", 2055), ("上下九", 5067), ("中山大学", 4244),
    ("华南师范大学", 1868), ("海珠桥", 3484),
    ("北京路", 1112), ("白云山", 980),
    ("广州融创", 865), ("芙蓉嶂", 847), ("火炉山", 678),
    ("臭屁醋", 1582), ("南越王", 868),
    ("赤坭皇家学院", 7555), ("猎德村民", 2550), ("干蒸烧卖", 2550)
]

w = WordCloud()
# 使用 add 方法添加数据，并设置词云图的标题
w.add("广州欢迎您！", words, word_size_range=[20, 80])
# 设置标题
w.set_global_opts(title_opts=opts.TitleOpts(title="词云图"))
# 渲染图表并保存为 HTML 文件
w.render("wordcloud.html")
w.render_notebook()
```
