```python
from pyecharts.charts import Pie
from pyecharts import options as opts

# 定义职称列表，包含高校教师的不同职称级别
L1 = ["教授", "副教授", "讲师", "助教", "其他"]
# 定义对应职称的人数数据列表
data = [20, 30, 50, 60, 15]

# 创建饼图对象
pie = Pie()

# 添加数据
pie.add(
    "职称", [list(z) for z in zip(L1, data)],
    radius=["40%", "75%"], rosetype=True
)

# 设置图表全局配置选项
# 设置图表标题为"某高校职称类别比例"
pie.set_global_opts(title_opts=opts.TitleOpts(title="某高校职称类别比例"))
# 设置数据标签的显示格式
# 使用"{b}: {c}"格式，其中{b}表示数据项名称，{c}表示数据值
pie.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))

# 渲染图表
pie.render("pie_base.html")
pie.render_notebook()
```
