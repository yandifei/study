```python
# 1、导入模块
from pyecharts.charts import Bar
from pyecharts import options as opts

#2、准备数据
x=["衬衫", "毛衣", "领带", "裤子", "风衣", "高跟鞋", "袜子"]
y1=[114, 55, 27, 101, 125, 27, 105]
y2=[120,80,10,50,180,94,200]

# 3、创建对象
bar = (Bar()
.add_xaxis(x)
.add_yaxis("商家A", y1)
.add_yaxis("商家B", y2)
.set_global_opts(title_opts=opts.TitleOpts(title="商品销售情况", subtitle="A公司和B公司"),
toolbox_opts=opts.ToolboxOpts(is_show=True)
))

bar.set_series_opts(label_opts=opts.LabelOpts(position='right'))
bar.reversal_axis()

#3、渲染图表
# 渲染图表
bar.render("bar.html")
bar.render_notebook()
```
