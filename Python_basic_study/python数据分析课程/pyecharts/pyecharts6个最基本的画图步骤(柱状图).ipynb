```python
# 1、导入模块
from pyecharts.charts import Bar
from pyecharts import options as opts

# 2、准备数据
x=["衬衫", "毛衣", "领带", "裤子", "风衣", "高跟鞋", "袜子"]
y=[114, 55, 27, 101, 125, 27, 105]

# 3、创建对象
bar = Bar()

# 4、把数据传入对象
bar.add_xaxis(x)
bar.add_yaxis("商品数量", y)

# 5、设置图标选项
bar.set_global_opts(
    title_opts=opts.TitleOpts(title="商品销售情况"),
    xaxis_opts=opts.AxisOpts(name="商品"),
    yaxis_opts=opts.AxisOpts(name="商家")
)

# 6、渲染图表
bar.render("bar.html")
bar.render_notebook()
```
