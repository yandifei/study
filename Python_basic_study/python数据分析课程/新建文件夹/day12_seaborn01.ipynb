# 主题/边框设置

```python
import seaborn as sns       #导入seaborn库
import numpy as np          #导入numpy库
import matplotlib.pyplot as plt #导入matplotlib.pyplot库
import pandas as pd
data = {
    '地区':['西南','华南', '华北','华中','东北', '西北', '华东'],
    '销售额':[138, 186, 226, 238,278,356,385],
    }
df = pd.DataFrame(data)

p1=df['地区']
p2=df['销售额']

plt.plot(df.index,p2,color="y",marker='d') 
sns.set_context('talk')
#sns.despine()#移除图表的上、右边框线，使图表更加简洁
#sns.despine(offset=20,trim=True)#移除图表的上、右边框线，并设置偏移量为20，修剪边框线
#sns.despine(left=True,bottom=True)#移除图表的左边框线使图表更加简洁
plt.show()
```

```python
fig=plt.figure(figsize=(12,8))

sns.set(style="darkgrid",font_scale=1.5)
ax1=fig.add_subplot(2,2,1)  
plt.plot(df.index,p2,color="g",marker='o') 

sns.set(style="dark")
ax2=fig.add_subplot(2,2,2) 
plt.plot(df.index,p2,color="m",marker='*') 

sns.set(style="whitegrid")
ax3=fig.add_subplot(2,2,3) 
plt.plot(df.index,p2,color="r",marker='D') 

sns.set(style="white")
ax4=fig.add_subplot(2,2,4) 
plt.plot(df.index,p2,color="y",marker='d') 

plt.show()
```

# 直方图和核密度曲线图

```python
tips=sns.load_dataset('tips')  #读入seaborn自带的小费数据
tips.head()
```

```python
#distplot(a,bins,hist,kde,rug)
#a 表示数据；
#bins 表示柱体的数量；None 时由 Freedman-Diaconis 规则自动计算。
#hist 表示是否显示柱体；默认true
#kde 表示是否绘制核密度曲线；默认true
#rug 表示是否在x轴上显示观测值竖线（边际地毯图）。默认false
fig,axes = plt.subplots(2,2,figsize=(12,8))   #子图.fig:外框，axes:内框
# 直方图嵌套密度曲线图， distplot( )为hist加强版
sns.histplot(tips['total_bill'],ax=axes[0,0],kde=True)  
#不带密度图, bins设置分段
sns.histplot(tips['total_bill'], ax=axes[0,1], bins=20, kde=False)
 # kdeplot( )为密度曲线图   shade  阴影  
sns.kdeplot(tips['total_bill'], ax=axes[1,0], fill=True, color='r')   
# rugplot 边际毛毯图
sns.rugplot(tips['total_bill'], ax=axes[1,1], color='g') 
plt.show()  
```

# 散点图

```python
n1=plt.figure(figsize=(12,8))
ax1=n1.add_subplot(2,2,1)  
plt.scatter(x=tips['day'],y=tips['tip'], c='r')  #matplotlib的散点图

ax2=n1.add_subplot(2,2,2) 
sns.stripplot(x='day', y='tip',data= tips,color='m')  #seaborn的散点图

ax2=n1.add_subplot(2,2,3) 
#散射点通常会重叠，jitter参数可以随机抖动离散点
sns.stripplot(x='day', y='tip',data= tips,color='c',jitter=True) 
ax2=n1.add_subplot(2,2,4) 
#使用 swarmplot(),避免散点重叠带分布的散点图，hue参数可以再嵌套一个变量
sns.swarmplot(x='day', y='tip',data= tips,hue='sex',size=3)  
plt.show()
```

# 直方图

```python
sns.boxplot(x=tips['day'],y=tips['total_bill'],data=tips)
plt.show()
```

# 提琴图

```python
sns.set(style='whitegrid', palette='Accent')
fig, axes = plt.subplots(1,2,figsize=(14, 5))  #子图
#split:小提琴是否分成两半，默认false
#inner：控制小提琴图内部显示的统计信息，可选值包括：'box': 显示微型箱线图（默认）。'quartiles': 显示四分位线。
#'point' / 'stick': 显示具体的数据点或茎叶图。None: 不显示内部细节，仅保留平滑的小提琴轮廓
sns.violinplot(x='day', y='total_bill', hue='time', data=tips,ax=axes[0])
#split参数可以将分类数据进行切分，琴形图两边的颜色代表不同的类别,inner参数对每个数据可视化
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips,split=True,inner='stick', ax=axes[1])
plt.show
```

# 条形图和点图

```python
#iris是鸢尾花卉是数据集，每个样本包含了花萼长度、花萼宽度、花瓣长度、花瓣宽度四个特征（前4列）
iris = sns.load_dataset('iris')  
iris.head()
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')  
fig, axes = plt.subplots(1,2,figsize=(14, 6))
sns.barplot(x='day', y='total_bill', data=tips,hue='sex',palette='Accent',ax=axes[0])
sns.countplot(x='petal_width', data=iris, hue='petal_width', palette='husl', legend=False, ax=axes[1]) 
plt.show()
```

```python
# 绘制条形图
sns.pointplot(x="day", y="total_bill", data=tips)
```

# 回归图

```python
sns.regplot(x="total_bill", y="tip",data=tips,ci=None) # ci设置阴影
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips,ci=None)  #增加一个变量
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker',col='time',row='sex',data=tips)  #再增加两个变量
```

# 多变量图

```python
tips=sns.load_dataset('tips')
sns.jointplot(x='total_bill', y='tip',data=tips)  
sns.jointplot(x='total_bill', y='tip',data=tips,kind='kde')  
sns.jointplot(x='total_bill', y='tip',data=tips,kind='hex')  
sns.jointplot(x="total_bill", y="tip",data=tips, kind='reg')
plt.show()
```

# 散点图矩阵

```python
#在 Seaborn 中使用  pairplot  方法实现数据特征的两两对比。默认是所有特征，可以通过  vars 参数指定部分特征。
#hue:分类变量
#‌vars‌：指定参与绘图的变量列表，避免图形过于密集，仅展示关心的特征列 。‌‌
#palette：颜色方案
sns.pairplot(iris, vars=["sepal_width", "sepal_length"],hue='species',palette="husl")
#sns.pairplot(iris,hue='species',palette="husl")
plt.show()
```
