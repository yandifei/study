# 重置索引

```python
data1 = pd.Series([1,2,3,4,5],index=['b','c','a','e','d'])
data1
```

```python
data1.reindex(['a','b','c','d','e'])
```

```python
data1.reindex(['a','b','c','d','e','f'])
```

```python
data1.reindex(['a','b','c','d','e','f','g'],fill_value=10)
```

```python
data2 = pd.Series([1,3,5,7],index=[0,2,4,6])
data2
```

```python
data2.reindex(range(6))
```

```python
data2.reindex(range(6),method='ffill')
```

```python
data2.reindex(range(6),method='bfill')
```
