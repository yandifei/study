```python
import pandas as pd
```

```python
# 导入pandas库后创建的两个Series对象
data1 = pd.Series(range(10,13),index=range(3))  # 创建一个Series，值为10到12，索引为0到2
data2 = pd.Series(range(20,25),index=range(5))  # 创建一个Series，值为20到24，索引为0到4
```

```python
data1
```

```python
data2
```

```python
# 将data1和data2进行相加操作
# 这是一个简单的数据相加表达式
data1+data2
```

```python
# 将data2的数据添加到data1中，如果遇到缺失值则用0填充
data1.add(data2, fill_value=0)
# 这个方法有缺失值填充，以后用这个，因为更灵活
```
