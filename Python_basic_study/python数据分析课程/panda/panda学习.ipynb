```python
# 导入 pandas 库，并简写为 pd（这是惯例用法）
import pandas as pd
```

```python
# 通过列表(List)创建简单的 Series
# 如果不指定索引，Pandas 会自动创建一个从 0 开始的整数索引
data = pd.Series([2,3,4,5])
data
```

```python
# 创建 Series 时手动指定自定义索引 (index)
# 使用 index 参数可以将默认的数字索引替换为标签（如 'a', 'b' 等）
data1 = pd.Series([2,3,4,5],index=['a','b','c','d'])
data1
```

```python
# 使用字典 (Dictionary) 创建 Series
dic = {"name": "zhangsan", "titles": "techer"}
data2 = pd.Series(dic)
data2
```

```python
# 检查变量data2的数据类型
type(data2)
```

```python
# 获取data2的索引
# index属性用于返回Series或DataFrame的索引对象
# 对于Series，它是一维标签数组
# 对于DataFrame，它是行标签的Index对象
data2.index
```

```python
# 获取data2中的所有值
# values属性用于返回Series或DataFrame中的数据值
# 对于Series，它返回一个NumPy数组，包含Series中的所有值
# 对于DataFrame，它返回一个二维NumPy数组，包含DataFrame中的所有值
data2.values
```

```python
# 访问列表data2中的第二个元素
# 列表的索引从0开始，因此[1]表示第二个元素
data2[1]
```

```python
# 使用iloc索引器获取data2 DataFrame中第二行（索引为1）的数据
# iloc是基于整数位置的索引，不考虑实际的行标签
data2.iloc[1]
```

```python
# 将变量 data 中的内容重复两次
data * 2
```

```python
# 将data2变量与字符串"ha"进行拼接操作
data2 + "ha"
```
