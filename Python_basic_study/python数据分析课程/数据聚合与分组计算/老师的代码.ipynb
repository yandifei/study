# 数据分组与聚合
## 数据分组

```python
import pandas as pd
import numpy as np
```

```python
#通过列名
df = pd.DataFrame({"Key":['C','B','C','A','B','B','A','C','A'],
                   "Data":[2,4,6,8,10,1,14,16,18]})
df
```

```python
group_obj = df.groupby(by='Key')
```

```python
for i in group_obj:
    print(i)
```

```python
#通过Serios对象分组
df = pd.DataFrame({'keyl': ['A', 'A', 'B', 'B', 'A'],
                   'key2': ['one','two','one', 'two', 'one'],
                    'datal': [2, 3, 4, 6,8],
                    'data2': [3, 5, 6, 3, 7]})
df
```

```python
se = pd.Series(['a','b','a','c','b'])
```

```python
group_obj = df.groupby(by=se)
```

```python
for i in group_obj:
    print(i)
```

```python
#通过字典
df = pd.DataFrame({'a': [1, 2, 3, 4, 5],
                   'b': [6, 7, 8, 9, 10],
                   'c': [11, 12, 13, 14, 15],
                   'd':[5,4,3,2,1],
                    'e':[10,9,8,7,6]})
df
```

```python
mapping={'a':'第一组','b':'第二组','c':'第一组',
        'd':'第三组','e':'第二组'}
mapping
```

```python
group_obj = df.groupby(by=mapping,axis=1)
```

```python
for i in group_obj:
    print(i)
```

```python
#按照函数分组
df = pd.DataFrame({'a': [1, 2, 3, 4, 5],
                   'b': [6, 7, 8, 9, 10],
                   'c': [5, 4, 3, 2, 1]},
                    index=['Sun', 'Jack', 'Alice', 'Helen', 'Job'])
df
```

```python
group_obj = df.groupby(len)
```

```python
for i in group_obj:
    print(i)
```

# 聚合统计
## 使用内置方法统计聚合数据

```python
df = pd.DataFrame({'key1': ['A', 'A', 'B', 'B','A'],
                    'key2': ['one','two','one','two','one'],
                    "datal": [2, 3, 4, 6, 8],
                    "data2": [3, 5, 2, 3,7]})
df
```

```python
df.groupby(by='key1').mean
```

## 面向列的聚合方法（agg）

```python
data_frame = pd.DataFrame (np.arange(36).reshape ((6, 6)),
                           columns=list('abcdef'))
data_frame['key'] = pd.Series(list('aaabbb'), name='key')
data_frame
```

```python
data_group = data_frame.groupby(by='key')#获取分组对象
data_group.groups['a'] #获取行名称
```

```python
data_frame.iloc[data_group.groups['a']] #获取行名称相对应的数据
```

```python
data_frame.iloc[data_group.groups['b']]
```

```python
data_group.agg(sum)#求和
```

```python
# 求极差值
def rang_data_group(arr):
    return arr.max() -arr.min()
```

```python
data_group.agg(rang_data_group)
```

```python
# 对一组数据用两种函数聚合
data_group.agg([rang_data_group,'sum'])
```

```python
data_group.agg([('极差值',rang_data_group),('求和','sum')])
```

```python
# 每列使用不同函数聚合
data_group.agg({'a':'sum','b':'mean','c':rang_data_group,'d':'std'})
```

## 其他分组级运算
### 数据转型（transform）

```python
df = pd.DataFrame({'a':[0,1,6,10,3],
                   'b':[1,2,7,11,4],
                   'c':[2,3,8,12,4],
                   'd':[3,4,9,13,5],
                   'e':[4,5,10,14,3],
                   'key':['A','A','B','B','B'],
})
df
```

```python
df.groupby(by='key').transform('mean')
```

```python
df=pd.DataFrame({'A':[2,3,3,4,2],
                 'B':[4,2,3,6,6],
                 'C':[9,7,0,7,8],
                 'D':[3, 4,8,6,10]})
df
```

```python
key=['one','one','two','two','two']
```

```python
df.groupby(by=key).transform('mean')
```

## 数据应用（apply）

```python
data_frame = pd.DataFrame({'datal': [80,23,25,63,94,92,99,92,82,99],
                            'data2': [41,87,58,68,72,89,60,42,53,65],
                            'data3': [30,78,23,66,16,59,20,23,24,40]})
data_frame
```

```python
data_frame.apply(lambda x:x.max() - x.min())
```
