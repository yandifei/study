```python
import pandas as pd
import numpy as np
```

```python
csv = pd.read_csv("运动员信息表.csv", encoding="ANSI")
print(csv)
```

```python
# 查询篮球分组信息
csv.groupby(by="项目").get_group("篮球")
```

```python
## 2. 查看男运动员分组
csv.groupby("性别").get_group("男")
```

```python
## 3. 查看男篮分组年龄、身高、体重三列数的极差值
# 先拿到男篮分组
data = csv.groupby("性别").get_group("男").groupby("项目").get_group("篮球")
# 使用apply方法对DataFrame的每一列应用一个lambda函数
mapping = {"年龄（岁）": lambda x: x.max() - x.min(),
           "身高(cm)": lambda x: x.max() - x.min(),
           "体重(kg)": lambda x: x.max() - x.min()
           }
# 该lambda函数计算每列的最大值与最小值之差，即每列的极差
data.agg(mapping)


# obj = csv[csv["性别"] == "男"][csv["项目"] == "篮球"]
# obj
```

```python
## 4. 添加体质指数BIM值=身高/体重^2
# csv["BIM"] = csv["身高(cm)"] / (csv["体重(kg)"] ** 2)
# csv
def bmi(height, weight):
    return (height/100) / (weight * weight)
# 该lambda函数拿到身高和体重求BIM值
csv["BIM"] = csv.apply(lambda x: bmi(x["身高(cm)"], x["体重(kg)"]), axis=1)
csv
```

```python
def BMI(df):
    weight = df["体重(kg)"]
    height = df["身高(cm)"]
    return height / (weight ** 2)
```
