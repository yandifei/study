## groupby方法将数据分组
groupby()方法将数据按照某些规则标准划分
不会对实际对象造成影响
标准：列名、函数、索引、字典、序列、或索引级别

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({"Key":['C','B','C','A','B','B','A','C','A'],
                   "Data":[2,4,6,8,10,1,14,16,18]})
df
```

```python
group_obj = df.groupby(by="Key")
```

```python
for i in group_obj:
    print(i)
```

```python
df = pd.DataFrame({'key1': ['A', 'A', 'B', 'B','A'],
                    'key2': ['one','two','one','two','one'],
                    "datal": [2, 3, 4, 6, 8],
                    "data2": [3, 5, 2, 3,7]})
df
```

```python
# 导入pandas库并创建一个Series对象
# Series是pandas中的一维标记数组，能够容纳任何数据类型
se = pd.Series(['a','b','a','c','d'])  # 使用列表创建一个包含4个元素的Series，索引默认为0-3的整数
```

```python
# 使用pandas的groupby方法对DataFrame进行分组
# group_obj: 分组后的对象，是一个GroupBy对象
# df: 输入的DataFrame数据结构
# by: 分组的依据，这里用变量se表示分组列名
group_obj = df.groupby(by=se)
```

```python
# 遍历group_obj对象中的每个元素
for i in group_obj:
    # 打印当前遍历到的元素i
    print(i)
```

```python
df = pd.DataFrame({'a': [1, 2, 3, 4, 5],
                   'b': [6, 7, 8, 9, 10],
                   'c': [5, 4, 3, 2, 1],
                   'd': [5, 4, 3, 2, 1],
                   'e': [10, 9, 8, 7, 6]
                })
```

```python
mapping = {'a': "第一组",
           'b': "第二组",
           'c': "第一组",
           'd': "第三组",
           'e': "第二组"}
```

```python
group_obj = df.groupby(by=mapping, axis=1)
```

```python
for name, group in group_obj:
    print(f"\n组名: {name}")
    print(group)
```

```python
# 按照函数分组
df = pd.DataFrame({'a': [1, 2, 3, 4, 5], #  创建列'a'，包含5个元素
                   'b': [6, 7, 8, 9, 10], #  创建列'b'，包含5个元素
                   'c': [5, 4, 3, 2, 1]}, #  创建列'c'，包含5个元素
                    index=['Sun', 'Jack', 'Alice', 'Helen', 'Job']) #  设置行索引
df #  显示创建的DataFrame
```

```python
# 定义一个函数，输入参数为字符串s，返回字符串s的长度
group_obj = df.groupby(len)
```

```python
# 遍历group_obj对象中的每个元素
for i in group_obj:
    print(i)
```

## 聚合统计
### 使用内置方法统计聚合数据

```python
df = pd.DataFrame({'key1': ['A', 'A', 'B', 'B','A'],
                    'key2': ['one','two','one','two','one'],
                    "datal": [2, 3, 4, 6, 8],
                    "data2": [3, 5, 2, 3,7]})
df
```

```python
# 使用pandas的groupby方法按照"key1"列进行分组，并计算每组的平均值
# numeric_only=True参数表示只计算数值列的均值，忽略非数值列
df.groupby("key1").mean(numeric_only=True)
```

## 面向列的聚合方法（agg）

```python
data_frame = pd.DataFrame (np.arange(36).reshape ((6, 6)),
                           columns=list('abcdef'))
data_frame['key'] = pd.Series(list('aaabbb'), name='key')
data_frame
```

```python
# 对DataFrame按照"key"列进行分组
data_group = data_frame.groupby("key")
# 获取分组中键为'a'的所有数据索引
data_group.groups['a']
```

```python
#  使用iloc方法根据分组结果获取特定组的数据 data_frame 是一个数据框对象 data_group 是一个分组对象，通过groups属性可以获取每个组的数据索引 'a' 是分组的键名，表示要获取的特定组 最终返回的是分组键为'a'的所有数据行
data_frame.iloc[data_group.groups['a']]
```

```python
# 使用iloc方法根据分组结果获取特定组的数据 data_frame 是一个数据框对象 data_group 是一个分组对象，通过groups属性可以获取每个组的数据索引 'a' 是分组的键名，表示要获取的特定组 最终返回的是分组键为'a'的所有数据行
data_frame.iloc[data_group.groups['a']]
```

```python
# 个警告 FutureWarning 的含义是：你当前传递给 agg 的参数是一个函数对象（sum），但在未来的 Pandas 版本中，这种行为将会改变。为了避免未来的代码报错，建议直接传递字符串 "sum"
data_group.agg("sum")
```

```python
# 求极差值
def range_dat_group(arr):
    return arr.max() - arr.min()
```

```python
data_group.agg(range_dat_group)
```

```python
# 对一组数据用两种函数聚合
data_group.agg([range_dat_group, "sum"])
```

```python
data_group.agg([("极差值", range_dat_group), ("总和", "sum")])
```

```python
# 每列使用不同的函数集合
data_group.agg({"a": range_dat_group, "b": "sum", "c": range_dat_group, "d": "sum"})
```

# 其他分组级运算
## 数据转型(transfer)

```python
df = pd.DataFrame({'a':[0,1,6,10,3],
                   'b':[1,2,7,11,4],
                   'c':[2,3,8,12,4],
                   'd':[3,4,9,13,5],
                   'e':[4,5,10,14,3],
                   'key':['A','A','B','B','B'],
})
df
```

```python
obj = df.groupby("key")
```

```python
for i in obj:
    print(i)
```

```python
# 使用pandas的groupby方法按"key"列对数据框df进行分组，然后应用transform方法计算每组的平均值
# transform方法会将计算结果广播回原始数据框的形状，即每个原始行都会显示其对应组的平均值
obj = df.groupby("key").transform("mean")
obj
```

```python
df=pd.DataFrame({'A':[2,3,3,4,2],
                 'B':[4,2,3,6,6],
                 'C':[9,7,0,7,8],
                 'D':[3, 4,8,6,10]})
df
```

```python
key = ['one', 'one', 'two', 'two', 'two']
```

```python
df.groupby(by=key).transform("mean")
```

## 数据应用（apply）

```python
data_frame = pd.DataFrame({'datal': [80,23,25,63,94,92,99,92,82,99],
                            'data2': [41,87,58,68,72,89,60,42,53,65],
                            'data3': [30,78,23,66,16,59,20,23,24,40]})
data_frame
```

```python
# 使用apply方法对DataFrame的每一列应用一个lambda函数
# 该lambda函数计算每列的最大值与最小值之差，即每列的极差
data_frame.apply(lambda x: x.max() - x.min())
```
