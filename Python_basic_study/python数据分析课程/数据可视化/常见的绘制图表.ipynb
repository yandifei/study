# 常见绘制图表

```python
import matplotlib.pyplot as plt
import numpy as np
# 设置字体为黑体以支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  
# 解决保存图像是负号'-'显示为方块的问题
plt.rcParams['axes.unicode_minus'] = False    
```

```python
# 导入所需的库（假设已经导入）
# numpy用于生成随机数，matplotlib.pyplot用于绘图
# 生成100个符合标准正态分布的随机数
data = np.random.randn(100)
# 绘制直方图
# data: 要绘制的数据
# bins: 将数据分成20个区间
# color: 设置直方图颜色为绿色(g)
# alpha: 设置透明度为0.5，使重叠部分可见
plt.hist(data, bins=20, color='g', alpha=0.5)
# 显示图形
plt.show()
```

```python
# 创建 x 坐标：0 到 50 的整数数组（共51个点）
x = np.arange(51)
# 生成 y 坐标：51个 0~1 之间的随机数（均匀分布）
# 注意：这里覆盖了之前用 arange 生成的 y，所以最终 y 是随机值
y = np.random.rand(51)

# 绘制散点图
plt.scatter(
    x,           # x轴数据
    y,           # y轴数据
    c='r',       # 颜色：红色 (red)
    marker='*',  # 标记样式：*五角星【^是三角形】
    s=100        # 点的大小（面积），数值越大点越大【这里指的是离散图像的像素】
)

# 显示图形
plt.show()
```

```python
# 创建x轴的数据，范围从0到4的整数
x = np.arange(5)
# 生成两组随机整数数据，范围在1到30之间，每组5个数据
y1, y2 = np.random.randint(1, 31, size=(2, 5))
# 创建一个1行1列的子图网格，并选择第一个子图进行绘制
ax = plt.subplot(1, 1, 1)
# 在图表中width默认值是0.5
width = 0.3
# 绘制第一组柱状图，颜色为红色，宽度为width，透明度为0.5
ax.bar(x, y1, width, color='r', alpha=0.5)
# 绘制第二组柱状图，颜色为绿色，宽度为width，透明度为0.5
# x坐标向右偏移width，使两组柱状图并排显示
ax.bar(x + width, y2, color='g', width=width, alpha=0.5)
# 设置 X 轴刻度和标签
# 将刻度设置在每组柱子的中间位置 (x + width/2)
ax.set_xticks(x + width/2)
# 将刻度标签设置为月份名称
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May'])
# ax.xticks(x + width / 2, labels=x)
plt.show()
```

```python
# 导入matplotlib库中的pyplot模块，用于绘图
plt.figure(figsize=(6,6))  # 设置图表大小为6x6英寸
# 定义饼图的标签，分别为春、夏、秋、冬四个季节
labels='Spring','Summer','Autumn','Winter'
# 定义每个部分的数据值，分别对应四季的销量
x = [15,30,45,10]
# 定义突出显示的部分，数值越大，突出程度越高
explode = (0.05,0.05,0.1,0.05)
plt.pie(x,labels=labels,startangle=90,explode=explode,autopct='%1.1f%%')
plt.title('某商品四季销量')
plt.show()
```
