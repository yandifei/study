### 常见图表类型汇总表

| 图表类型 | 定义 | 主要用途与功能 | 特性与优缺点 |
| --- | --- | --- | --- |
| **直方图** | 由一系列高度不等的纵向条纹或线段表示数据分布的图形，又称质量分布图。 | 适于比较数据之间的多少，显示数据分布情况。 | 利用方块高度显示差异；仅适用于中小规模数据集。 |
| **折线图** | 用直线段将各数据点连接起来而组成的图形。 | 显示数据的变化趋势，反映一组数据的动态。 | 适合显示随时间变化的连续数据，通常要求时间间隔相等。 |
| **条形图** | 用宽度相同但长短/高度不同的条形表示数据多少的图形。 | 显示各个项目之间的比较情况，其作用与直方图类似。 | 条形可横置或纵置（纵置时称为柱形图），对比效果直观。 |
| **饼图** | 显示一个数据序列中各项大小与总和比例的图形。 | 清楚地反映出各数据系列的百分比情况。 | 每项具有唯一颜色并与图例对应；强调部分与整体的关系。 |
| **散点图** | 数据点在直角坐标系平面上的分布图。 | 显示若干数据系列中各数值间的关系，判断变量相关性。 | 数据点越多比较效果越好；用于表示因变量随自变量变化的趋势。 |
| **箱形图** | 又称盒须图或箱线图，显示一组数据分散情况的统计图。 | 识别异常值方面有一定优越性，用于简单总结数据。 | 仅用 5 个统计点（如中位数、极值等）即可总结数据分布。 |

### 核心功能总结

* **趋势分析**：首选**折线图**，用于观察数据随时间的波动。
* **数据对比**：**条形图**和**直方图**能最直观地展示不同类别或项目间的数值大小差异。
* **占比分布**：**饼图**最擅长表达各组成部分在总量中的百分比。
* **相关性与离散度**：通过**散点图**研究变量间的关联，利用**箱形图**快速掌握数据的离散程度及异常值。

```python
# 导入必要的库（假设已经导入numpy和matplotlib.pyplot）
import numpy as np
import matplotlib.pyplot as plt
# 创建一个从0到100的数组
data = np.arange(0, 101)
# 创建2x2的子图布局，第一个子图位于左上角
plt.subplot(221)
# 绘制data与自身的线性关系图
plt.plot(data, data)
# 创建2x2的子图布局，第二个子图位于右上角
plt.subplot(222)
# 绘制data与-data的线性关系图
plt.plot(data, -data)
# 创建2x1的子图布局，第三个子图占据下方整个区域
plt.subplot(212)
# 绘制data的平方图
plt.plot(data, data**2)
plt.show()
```

subplots() 函数会返回一个元组，元组的第一个元素为 Figure 对象（画布），第二个元素为 Axes 对象（子图，包含坐标轴和画的图）或 Axes 对象数组。 如果创建的是单个子图，则返回的是一个 Axes 对象，否则返回的是一个 Axes 对象数组。

```python
# 创建一个从0到100的数组
data = np.arange(0, 101)  # 使用arange函数生成0到100的整数序列

# 创建一个2行2列的图表布局，返回图表和坐标轴对象的元组
fig, axes = plt.subplots(2,2)  # subplots函数用于创建多个子图，2行2列共4个子图
ax1 = axes[0,0]  # 获取左上角的子图
ax2 = axes[0,1]  # 获取右上角的子图
ax3 = axes[1,0]  # 获取左下角的子图
ax4 = axes[1,1]  # 获取右下角的子图

# 在每个子图上绘制数据
# 在左上角子图上绘制data与自身的线性关系图
ax1.plot(data, data)
# 在第二个子图上绘制 data 和 -data 的关系曲线
ax2.plot(data, -data)
# 在第三个子图上绘制 data 和 data 的平方的关系曲线
ax3.plot(data, data ** 2)
# 在第四个子图上绘制 data 的正值与其自然对数的关系曲线
# 注意这里只选取了 data 中大于 0 的值进行绘制，因为对数函数的定义域为正实数
ax4.plot(data[data > 0], np.log(data[data > 0]))
plt.show()  # 显示图表
```

通过 add_subplot() 函数添加和选中子图

```python
# 导入必要的库（假设已经导入，因为代码中没有显示导入语句）
# matplotlib.pyplot 用于绘图
# numpy 用于生成随机数据
# 创建一个新的图形窗口
fig = plt.figure()
# 在图形窗口中添加子图
# 参数2,2表示创建一个2行2列的子图网格
# 第一个参数1表示选择第一个子图（左上角）
fig.add_subplot(2, 2, 1)
# 选择第二个子图（右上角）
fig.add_subplot(2, 2, 2)
# 选择第四个子图（右下角）
fig.add_subplot(2, 2, 4)
# 选择第三个子图（左下角）
fig.add_subplot(2, 2, 3)
# 生成100个标准正态分布的随机数
data = np.random.randn(100)
# 绘制数据线图
# 注意：此绘图命令会使用最后一个活动的子图（即左下角的子图）
plt.plot(data)
# 显示图形窗口
plt.show()
```

添加各类标签

```python
# 设置中文字体
# 用来正常显示中文标签
plt.rcParams['font.sans-serif'] = ['SimHei']    # 黑体
# 用来正常显示负号
plt.rcParams['axes.unicode_minus'] = False


# 创建一个新的图形窗口
fig = plt.figure(figsize=(16, 8))  # 设置图形大小
fig.add_subplot(1,2,1)

# 导入必要的库（假设已经导入numpy和matplotlib.pyplot）
data1 = np.arange(0,1,0.1)  # 创建一个从0到1，步长为0.1的数组

# 设置图表标题和坐标轴标签
plt.title("二次函数和三次函数")  # 设置图表标题为"二次函数和三次函数"
plt.xlabel("x轴")  # 设置x轴标签为"x轴"
plt.ylabel("y轴")  # 设置y轴标签为"y轴"

# 设置x轴的显示范围
plt.xlim(0,0.5)  # x轴的范围从0到0.5
# 设置y轴的显示范围
plt.ylim(0,0.5)  # y轴的范围从0到0.5

# 设置x轴和y轴的刻度
plt.xticks([0,0.5,1])  # 设置x轴刻度为0、0.5和1
plt.yticks([0,0.5,1])  # 设置y轴刻度为0、0.5和1


# 导入matplotlib库中的pyplot模块，并简写为plt
import matplotlib.pyplot as plt
# 绘制第一条曲线，x轴为data数据，y轴为data的平方
plt.plot(data1,data1**2)
# 绘制第二条曲线，x轴为data数据，y轴为data的立方
plt.plot(data1,data1**3)
# 添加图例，说明第一条曲线是"x=y^2"，第二条曲线是"x=y^3"
plt.legend(["x=y^2","x=y^3"])


# 在 1x2 的布局中添加并选中第 2 个子图
fig.add_subplot(1,2,2) 

# 准备数据：生成 0 到 2π 之间步长为 0.01 的数组
data2 = np.arange(0, np.pi*2, 0.01)

# 设置当前子图的标题
plt.title('正弦函数和余弦函数')

# 设置 x 轴和 y 轴的标签名称
plt.xlabel('X轴')
plt.ylabel('Y轴')

# 指定 x 轴刻度的取值（从 0 到 2π 的关键点）
plt.xticks([0, np.pi/2, np.pi, np.pi*3/2, np.pi*2])

# 指定 y 轴刻度的取值范围及间隔
plt.yticks([-1, -0.5, 0, 0.5, 1])

# 绘制正弦曲线
plt.plot(data2, np.sin(data2))

# 绘制余弦曲线
plt.plot(data2, np.cos(data2))

# 添加图例，区分 sin 和 cos 曲线
plt.legend(['sin', 'cos'])

# 显示图形
plt.show()
```

## 颜色、线型、标记设置

```python
# 导入必要的库（假设已导入numpy和matplotlib.pyplot）
import numpy as np
import matplotlib.pyplot as plt
# 创建一个从1开始，到3结束（不包括3），步长为0.3的数组
data = np.arange(1,3,0.3)
# 绘制第一条线：红色，虚线，圆形标记点
plt.plot(data, color='r', linestyle='--', marker='o')
# 绘制第二条线：黄色，实线，星形标记点
plt.plot(data+1, color='y', linestyle='-', marker='*')
# 绘制第三条线：绿色，点线，三角形标记点
plt.plot(data+2, color='g', linestyle=':', marker='^')
# 将图形保存为名为"picture"的文件
plt.savefig("picture")  #  必须在show之前，因为show会清空画布
# 显示图形
plt.show()
```
