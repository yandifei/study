```python
from pyecharts import options as opts
from pyecharts.charts import Tab, Line, Bar

# 数据准备
x = ['小米', '三星', '华为', '苹果', '魅族', 'OPPO', 'VIVO']
y1 = [100, 125, 87, 90, 78, 98, 118]
y2 = [78, 95, 120, 102, 88, 108, 98]

# 1. 配置柱状图
bar = (
    Bar()
    .add_xaxis(x)
    .add_yaxis('商家A', y1)
    .add_yaxis('商家B', y2)
    .set_global_opts(title_opts=opts.TitleOpts(title='组合图表-柱状图'))
)

# 2. 配置折线图
line = (
    Line()
    .add_xaxis(x)
    .add_yaxis('商家A', y1)
    .add_yaxis('商家B', y2)
    .set_global_opts(title_opts=opts.TitleOpts(title='组合图表-折线图'))
)

# 3. 使用 Tab 容器组合
tab = (
    Tab()
    .add(bar, '柱状图')  # 第一个参数是图表对象，第二个是标签名
    .add(line, '折线图')
)

# 渲染生成 HTML 文件
tab.render('tab.html')
tab.render_notebook()  # 在 Jupyter Notebook 中显示
```
