```python
# 导入 pyecharts 用于绘制 K 线图的类和配置项
from pyecharts.charts import Kline
from pyecharts import options as opts

# 导入数据处理库 pandas 并简写为 pd
import pandas as pd
# 导入数值计算库 numpy 并简写为 np
import numpy as np

# 创建一个名为 data 的字典，用于存放时间序列和模拟的开盘价数据
data = {
    # 生成从 2026-01-01 到 2026-01-20 的连续日期序列（共 20 天）
    'date': pd.date_range(start='2026-01-01', end='2026-01-20'),
    
    # 使用正态分布（高斯分布）随机生成 20 个开盘价数据
    # np.random.normal(loc, scale, size) 的参数含义：
    # loc=100: 正态分布的均值（数据的中心点在 100 左右）
    # scale=5: 正态分布的标准差（数据的波动范围，标准差越大波动越大）
    # size=20: 生成随机数的数量（正好对应上面的 20 天）
    'open': np.round(np.random.normal(100, 5, 20), 2),
    'close': np.round(np.random.normal(105, 7, 20), 2),
    'high': np.round(np.random.normal(110, 6, 20), 2),
    'low': np.round(np.random.normal(90, 4, 20), 2)
}

# 导入pandas库并创建DataFrame对象
df = pd.DataFrame(data)

# 创建Kline（K线图）对象实例
k = Kline()

# 为K线图添加X轴数据，将DataFrame中的'date'列转换
k.add_xaxis(df['date'].dt.strftime('%Y-%m-%d').tolist())
# 为K线图添加Y轴数据，将DataFrame中的'open', 'close', 'high', 'low'列转换为列表形式
k.add_yaxis("股价", y_axis=df[['open', 'close', 'high', 'low']].values.tolist())


# 设置全局配置项
k.set_global_opts(
    # 设置标题
    title_opts=opts.TitleOpts(title="K线图示例"),
)

# 渲染K线图并保存为html文件
k.render("render.html")
k.render_notebook()
```
