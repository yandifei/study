```python
# 从 pyecharts 的 charts 模块中导入 Gauge 类，用于创建仪表盘图
from pyecharts.charts import Gauge
from pyecharts import options as opts

# 初始化一个 Gauge 仪表盘图实例
g = Gauge()

# 向仪表盘实例中添加数据并进行更规范的样式配置
# '业务指标': 系列名称
# [('完成率', 55.5)]: 数据对，表示当前指标值为 55.5
# axisline_opts=opts.AxisLineOpts(...): 坐标轴轴线配置项
# linestyle_opts=opts.LineStyleOpts(...): 嵌套在线条配置项中，用以控制具体线条的颜色、宽度、阴影等样式
#   - color=[('red', 0.3), ('green', 0.4), ('purple', 0.3)]: 
#     将轴线按比例分段着色：
#     * 前 30% 区间显示为 红色 ('red')
#     * 中间 40% 区间显示为 绿色 ('green')
#     * 最后 30% 区间显示为 紫色 ('purple')
#   - width=20: 设置仪表盘圆弧轴线的宽度（线条粗细）为 20 像素
g.add(
    '业务指标', 
    [('完成率', 55.5)],
    axisline_opts=opts.AxisLineOpts(
        linestyle_opts=opts.LineStyleOpts(
            color=[('red', 0.3), ('green', 0.4), ('purple', 0.3)], 
            width=20
        )
    )
)

# 设置仪表盘的全局配置项，指定主标题为 'Gauge-不同颜色'
g.set_global_opts(title_opts=opts.TitleOpts(title='Gauge-不同颜色'))

# 将图表渲染并保存为本地的 HTML 文件，文件名为 'gauge.html'
g.render('gauge.html')
g.render_notebook()
```
