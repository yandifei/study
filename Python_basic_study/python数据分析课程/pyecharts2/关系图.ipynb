```python
# 从 pyecharts 的 charts 模块中导入 Graph 类，用于创建关系图（网络拓扑图）
from pyecharts.charts import Graph
from pyecharts import options as opts

# 定义关系图的节点数据（名称和大小）以及节点之间的连线数据（起点和终点）
points = [
    {'name': 'A', 'symbolSize': 50},
    {'name': 'B', 'symbolSize': 40},
    {'name': 'C', 'symbolSize': 30},
    {'name': 'D', 'symbolSize': 20},
    {'name': 'E', 'symbolSize': 10}
]

lines = [
    {'source': 'A', 'target': 'B'},
    {'source': 'B', 'target': 'C'},
    {'source': 'B', 'target': 'D'},
    {'source': 'C', 'target': 'E'},
    {'source': 'D', 'target': 'E'}
]

G = Graph()
# 将刚才定义的节点和连线数据传入关系图实例中，设置标题并渲染为网页文件
G.add('结点', points, lines, repulsion=5000)

G.set_global_opts(title_opts=opts.TitleOpts(title='基本关系网图'))

G.render('graph.html')
G.render_notebook()
```
