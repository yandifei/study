```python
# 从 pyecharts 的 charts 模块中导入 Funnel 类，用于创建漏斗图
from pyecharts.charts import Funnel
from pyecharts import options as opts

# 定义漏斗图各个环节的数据值列表
data = [45, 86, 39, 52, 68]
# 定义与数据值一一对应的标签（名称）列表
labels = ['电脑', '手机', '彩电', '冰箱', '洗衣机']

# 初始化一个 Funnel 漏斗图实例
f = Funnel()

# 向漏斗图实例中添加数据
# 参数 1: '电器销量图' 是该数据系列的名称（series_name）
# 参数 2: [list(z) for z in zip(labels, data)] 使用列表推导式和 zip 函数，
#        将标签和数据组合成 pyecharts 要求的格式，即：[['电脑', 45], ['手机', 86], ...]
f.add('电器销量图', [list(z) for z in zip(labels, data)])

# 设置漏斗图的全局配置项（注意：在使用 opts 之前，代码上方通常需要 import pyecharts.options as opts）
# title_opts=opts.TitleOpts(title='电器销量图') 用于设置图表的主标题为“电器销量图”
f.set_global_opts(title_opts=opts.TitleOpts(title='电器销量图'))

# 将生成的图表渲染并保存为本地的 HTML 文件，文件名为 'funnel.html'
# 运行后在同级目录下双击该 HTML 文件即可在浏览器中查看交互式漏斗图
f.render('funnel.html')
f.render_notebook()
```
