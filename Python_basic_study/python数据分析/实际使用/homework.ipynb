```python
# 切片与索引练习
# 2.1 创建0-24的5x5数组并索引
ar = np.arange(25).reshape(5, 5)
# 打印整个数组ar
print(ar)
# 打印数组ar中索引为4的元素
print(ar[4])
# 打印数组ar中前两行且从第三列开始到最后的所有元素
print(ar[:2, 3:])
# 打印数组ar中第四行第三列的元素
print(ar[3][2])
```

```python
# 统计分析
"""
班里有10名同学，随机生成成绩
现在需要用numpy来统计下这些人在python、网络爬虫、大学语文中的
平均成绩、最小成绩、最大成绩、方差、标准差，进行成绩输出。
"""
# 我假设成绩是0-100分，10个学生，每个学生3门成绩就是一个10x3的数组
scores = np.random.randint(0, 100, size=(10, 3))
scores
```

```python
# 定义这个数组的数据类型
# scores_dytype = np.dtype(
#     ("Python", int),
#     ("网络爬虫", int),
#     ("大学语文", int)
# )
```

```python
# # 将现在的类型转换为自定义的类型
# scores = scores.astype(scores_dtype)
# scores
```

```python
# 计算每门课的统计指标
# 计算每门课程成绩的平均值，axis=0表示按列计算
mean_scores = np.mean(scores, axis=0)
# 计算每门课程成绩的最小值，axis=0表示按列计算
min_scores = np.min(scores, axis=0)
# 计算每门课程成绩的最大值，axis=0表示按列计算
max_scores = np.max(scores, axis=0)
# 计算每门课程成绩的方差，axis=0表示按
var_scores = np.var(scores, axis=0)    
# 计算每门课程成绩的标准差，axis=0表示按列计算
std_scores = np.std(scores, axis=0)

# 输出统计结果
subjects = ["Python", "网络爬虫", "大学语文"]
for i, sub in enumerate(subjects):  # 遍历每门课程及其索引
    print(f"{sub}:")  # 输出课程名称
    print(f"  平均分: {mean_scores[i]}")  # 输出平均分，保留两位小数
    print(f"  最低分: {min_scores[i]}")      # 输出最低分
    print(f"  最高分: {max_scores[i]}")      # 输出最高分
    print(f"  方差: {var_scores[i]}")    # 输出方差，保留两位小数
    print(f"  标准差: {std_scores[i]}")  # 输出标准差，保留两位小数
```
