```python
import numpy as np
import csv
```

```python
# 导入必要的库（虽然代码中未显示，但使用csv.reader需要导入csv库）
# 创建一个空列表用于存储鸢尾花数据
iris_data = []
# 使用with语句打开文件，确保文件在使用后被正确关闭
# 'iris.csv'是要读取的文件名
# 'r'表示以只读模式打开文件
# encoding="gbk"指定文件编码为gbk格式
with open('iris.csv','r', encoding="gbk") as csvfile:
    # 创建CSV阅读器对象，用于逐行读取CSV文件内容
    csv_reader = csv.reader(csvfile)
    # 跳过CSV文件的标题行（第一行）
    print(csvfile.readline())
    # 遍历CSV阅读器中的每一行
    for row in csv_reader:
        # 将每一行数据转换为元组，并添加到iris_data列表中
        iris_data.append(tuple(row))
# 打印处理后的鸢尾花数据列表
print(iris_data)
# #数据清洗，将列表改成元祖，保证数据的只读性，不被篡改
# iris_list = []
# for row in iris_data:
#     iris_list.append(tuple(row))
# iris_list
```

```python
# 定义结构化数据类型，格式为np.dtype([(字段名1,类型1,长度1),(字段名2,类型2,长度2),...])，其中'<f8'表示浮点数，'<U10'表示字符串，'class'列的字符串长度为10
iris_type = np.dtype([
    ('Sepal.Length', np.str_, 40), # 字段名, 字符串类型, 最大长度40
    ('Sepal.Width',  np.str_, 40),
    ('Petal.Length', np.str_, 40),
    ('Petal.Width',  np.str_, 40),
    ('Species',      np.str_, 40)
])
```

```python
# 普通数组像只有编号的寄存柜，存取全靠数索引；结构化数组像贴了标签的药柜，能给每列定名字和类型，存取直接喊字段名。
iris_data = np.array(iris_data,dtype=iris_type)
iris_data
```

```python
Sepal_Length = iris_data['Sepal.Length'].astype(float)
Sepal_Length
```

```python
# 排序：将数值从小到大排列
np.sort(Sepal_Length)
```

```python
# 去重：提取数组中所有不重复的数值
np.unique(Sepal_Length)
```

```python
# 求和：计算数组中所有数值的总和
np.sum(Sepal_Length)
```

```python
# 累积和：返回一个数组，每个元素是前面所有元素的总和
np.cumsum(Sepal_Length)
```

```python
# 均值：计算平均数
np.mean(Sepal_Length)
```

```python
# 标准差：反映数据的离散程度（数值越小说明数据越集中）
np.std(Sepal_Length)
```

```python
# 方差：标准差的平方
np.var(Sepal_Length)
```
