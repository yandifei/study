# 重置索引

```python
# 导入pandas库并创建一个Series对象
data1 = pd.Series([1, 2, 3, 4, 5], index=["b", "c", "a", "e", "d"])  # 使用自定义索引创建Series，值为[1,2,3,4,5]，索引为["b","c","a","e","d"]
data1  # 输出Series对象data1
```

```python
# 使用reindex方法对数据进行重新索引
# reindex()方法用于根据新的索引标签重新排列数据
# 如果新索引标签在原数据中不存在，则填充NaN值
# 此处没有传入参数，表示使用默认行为，通常保持原有索引不变
data1.reindex()
```

```python
# 使用reindex方法对data1进行索引重排
data1.reindex(["a", "b", "c", "d", "e", "f"])  # 将data1的索引重新排列为指定的列表顺序
```

```python
# 使用reindex方法对data1进行重新索引，并指定填充值为10
data1.reindex(["a", "b", "c", "d", "e", "f"], fill_value=10)
```

```python
# 创建一个pandas Series对象
# 使用列表[1,3,5,7]作为数据
# 指定自定义索引[0,2,4,6]
pd.Series([1,3,5,7], index=[0,2,4,6])
```

```python
# 导入pandas库后创建的Series示例
# 使用pd.Series函数创建一个Series对象
data2 = pd.Series([1,3,5,7], index=[0,2,4,6])  # 创建一个包含值为[1,3,5,7]的Series，并指定索引为[0,2,4,6]
data2  # 输出Series对象，显示其值和索引
```

```python
# 使用reindex方法对data2进行重新索引，范围是0到5的整数
data2.reindex(range(6))
```

```python
# 使用reindex方法对data2进行重新索引
# range(6)生成索引范围0到5
# method="ffill"表示使用前向填充方法，即用前面的值填充缺失值
data2.reindex(range(6), method="ffill")
```

```python
# 使用reindex方法对data2进行重新索引，并使用method参数指定填充方法
# range(6)生成一个从0到5的整数序列作为新的索引
# method="bfill"表示使用后向填充（backward fill）方法，即用后面有效值填充缺失值
data2.reindex(range(6), method="bfill")
```
